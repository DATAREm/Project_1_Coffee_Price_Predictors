```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import matplotlib.pyplot as pltABC
import seaborn as sns
# from pandas import DataFrame, Series

Coffee_data = "Coffee_commodity_USDA.csv"

# Read Data File and store into Pandas Data Frames
Coffee_data_df = pd.read_csv(Coffee_data)
Coffee_data_df.head()
```

```python
reduced_Coffee_data_df = Coffee_data_df.loc[:, ["Country_Name", "Market_Year", "Value"]]
reduced_Coffee_data_df.head()
```

```python
reduced_Coffee_data_df = reduced_Coffee_data_df.loc[(
    reduced_Coffee_data_df["Value"] > 0) & (reduced_Coffee_data_df["Market_Year"] > 2013)]
reduced_Coffee_data_df.head()
```

```python
reduced_Coffee_data_df_ind = reduced_Coffee_data_df.reset_index(drop=True)
reduced_Coffee_data_df_ind.head()
```

```python
# reduced_Coffee_data_df_ctr_new = reduced_Coffee_data_df_ctr.groupby("Country_Name")["Value"].sum()
reduced_Coffee_data_df_ctr_new = reduced_Coffee_data_df_ind.groupby("Country_Name")

final_df = pd.DataFrame(reduced_Coffee_data_df_ctr_new["Value"].sum())
final_df.head()
```

```python
final_df = final_df.rename(
    columns={"Value": "Total Production"})
final_df.head()
```

```python
final_df = final_df.reset_index()
final_df.head()
```

```python
countries = ["Brazil", "Columbia", "Ethiopia", "Honduras","Peru"]
total_prod = [243800, 82875, 40428, 37160, 23650]
colors = ["brown","tan","sienna","goldenrod","chocolate"]
explode = (0.1,0,0,0,0)
plt.title("Coffee Production in Top 5 Countries in last 5 years")
plt.pie(total_prod, explode=explode, labels=countries, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.savefig("Coffee_predict.png")
plt.show()
```

```python
coffee_yearly_data = reduced_Coffee_data_df_ind.copy()
coffee_yearly_data.head()
```

```python
coffee_yearly_data_ind = coffee_yearly_data.set_index('Country_Name')
```

```python
a = coffee_yearly_data.loc[(coffee_yearly_data["Country_Name"] == 'Brazil')]
b = coffee_yearly_data.loc[(coffee_yearly_data["Country_Name"] == 'Honduras')]
c = coffee_yearly_data.loc[(coffee_yearly_data["Country_Name"] == 'Peru')]
d = coffee_yearly_data.loc[(coffee_yearly_data["Country_Name"] == 'Ethiopia')]
e = coffee_yearly_data.loc[(coffee_yearly_data["Country_Name"] == 'Colombia')]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
a.plot(kind="line", x="Market_Year",y="Value",grid=True, color = "brown",figsize=(10, 7),
              title="Country-wise Coffee Production", label = "Brazil", ax=ax)
b.plot(kind="line", x="Market_Year",y="Value",grid=True, color = "black",figsize=(10, 7),
              title="Country-wise Coffee Production", label = "Honduras", ax=ax)
c.plot(kind="line", x="Market_Year",y="Value",grid=True, color = "goldenrod",figsize=(10, 7),
              title="Country-wise Coffee Production", label = "Peru", ax=ax)
d.plot(kind="line", x="Market_Year",y="Value",grid=True, color = "chocolate",figsize=(10, 7),
              title="Country-wise Coffee Production", label = "Ethiopia", ax=ax)
e.plot(kind="line", x="Market_Year",y="Value",grid=True, color = "sienna",figsize=(10, 7),
              title="Country-wise Coffee Production", label = "Columbia", ax=ax)

plt.show()
# plt.tight_layout()
plt.savefig("Yearly_Coffee.png")
```
